```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_excel('../data/pd_films_gzdy_1.xlsx')
df
```

```python
n = 0
for i in df['Douban_higher']:  # convert str into machine-readable number
    if i == 'Yes':
        df['Douban_higher'][n]=1
    else:
        df['Douban_higher'][n]=0
    n = n+1
```

```python
df_1 = df[df['budget'] != 'no_budget_found']  # delete the items without budget
df_1
```

plt.style = {_classic_test_patch, seaborn-darkgrid, seaborn-ticks, fivethirtyeight, seaborn-whitegrid, classic, fast, seaborn-talk, seaborn-dark-palette, seaborn-bright, seaborn-pastel, grayscale, seaborn-notebook, ggplot, seaborn-colorblind, seaborn-muted, seaborn, Solarize_Light2, seaborn-paper, bmh, tableau-colorblind10, seaborn-white, dark_background, seaborn-poster, seaborn-deep}

```python
# make the data
plt.style.use('seaborn-whitegrid')

x = df_1['Douban_rate']
yy = df_1['boxoffice']*0.00001
sizes = df_1['budget']*0.00001

y = np.array(yy, dtype=np.float64)
size = np.array(sizes, dtype=np.float64)

plt.scatter(x, y, s=size, alpha=0.2,
           cmap='viridis')
plt.xlabel('Rating on Douban')
plt.ylabel('Box Office (US dollar/million)')
plt.show()
```

```python
# make the data
plt.style.use('seaborn-whitegrid')

#colors = np.random.uniform(15, 80, len(x))
x = df_1['IMDB_rate']
yy = df_1['boxoffice']*0.000001
sizes = df_1['budget']*0.00001
#color = df_1['Douban_higher']
df_1['year'] = np.abs(df_1['year']) * 100 +2

y = np.array(yy, dtype=np.float64)
size = np.array(sizes, dtype=np.float64)

plt.scatter(x, y, s=size, c=df_1['year'], alpha=0.2,
           cmap='viridis')
plt.xlabel('Rating on IMDb')
plt.ylabel('Box Office (US dollar/million)')
plt.axvline(6, color='r', linestyle='--', linewidth=1)
plt.axhline(50, color='r', linestyle='--', linewidth=1)
plt.legend()
plt.savefig('Rating on IMDb.png',dpi=600)
plt.show()
```
