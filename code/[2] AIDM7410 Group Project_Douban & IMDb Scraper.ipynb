# AIDM7410 Group Project_Douban & IMDb Scraper

This set of code is used to scrape the information of each movies from Douban and IMDb. It will cost about 2h to run this set of code. 

```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
import numpy as np
from time import sleep
from time import time
from random import randint
import re
```

## Douban Scraper

```python
title = []
DB_url = []
title = []
rate = []
reviews = []
year = []
type_ = []
imdbid = []


for y in range(2002, 2021):
    df = pd.read_excel('../data/film_xlsx/film_'+str(y)+'.xlsx')
    print(y)
    df_1 = df[df['Abandon']==" "] 
    for i in df_1['title']:
        title.append(i)
    n = 1
    for u in df_1['URL_Douban']: # extract url from excel file
        #DB_url.append(u)
        
        # Pause the loop
        sleep(randint(3,6))
        
        headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
        url = u
        page = requests.get(url, headers = headers)
        html_soup = BeautifulSoup(page.content, 'html.parser')
        container = html_soup.find('h1')
        
        ## rate
        try: 
            rate_1 = html_soup.find('div', id='interest_sectl')
            rate_2 = rate_1.find('strong').get_text()
            rate.append(rate_2)
        except: 
            rate_1 = 'no_rate_found'
            rate.append(rate_1) 

        ## reviews
        try: 
            reviews_1 = html_soup.find('div', class_='rating_sum')
            reviews_2 = reviews_1.find('span').get_text()
            reviews.append(reviews_2)
        except: 
            reviews_1 = 'no_reviews_found'
            reviews.append(reviews_1) 

        ## year
        try: 
            year_1 = container.find('span', class_='year').get_text()
            year.append(year_1[1:5])
        except: 
            year_1 = 'no_year_found'
            year.append(year_1) 

        ## type
        try:
            type_1 = html_soup.find_all('span', property='v:genre')
            if len(type_1) == 1:
                type1 = type_1[0].get_text()
                type_.append(str(type1))
            elif len(type_1) == 2:
                type1 = type_1[0].get_text()
                type2 = type_1[1].get_text()
                type_.append(str(type1)+' '+str(type2))
            elif len(type_1) == 3:
                type1 = type_1[0].get_text()
                type2 = type_1[1].get_text()
                type3 = type_1[2].get_text()
                type_.append(str(type1)+' '+str(type2)+' '+str(type3))
            elif len(type_1) == 4:
                type1 = type_1[0].get_text()
                type2 = type_1[1].get_text()
                type3 = type_1[2].get_text()
                type4 = type_1[3].get_text()
                type_.append(str(type1)+' '+str(type2)+' '+str(type3)+' '+str(type4))
            else:
                type1 = type_1[0].get_text()
                type2 = type_1[1].get_text()
                type3 = type_1[2].get_text()
                type4 = type_1[3].get_text()
                type5 = type_1[4].get_text()
                type_.append(str(type1)+' '+str(type2)+' '+str(type3)+' '+str(type4)+' '+str(type5))
        except: 
            type_1 = 'no_type_found'
            type_.append(type_1)


        ## IMDb id
        try:
            imdb = html_soup.find('div', id='info')
            imdb_1 = re.findall(r"<a class=\"ipc-metadata-list-item__icon-link\" href=\"(.*)\" aria-label=\"See full cast and crew\" target>",str(imdb))[0]
            imdbid.append(imdb_1)
        except: 
            imdb_1 = 'no_imdbid_found'
            imdbid.append(imdb_1)
        
        DB_url.append(u)
        print(n)
        n = n+1

print(len(title))
print(len(DB_url))
print(len(title))
print(len(rate))
print(len(reviews))
print(len(year))
print(len(type_))
print(len(imdbid))
```

```python
films_info = {
    'title': title,
    'year': year, 
    'DB_url': DB_url,
    'DB_rate': rate,
    'DB_reviews':reviews,
    'DB_type':type_,
    'IMDB_id':imdbid
}
```

```python
pd_films_info = pd.DataFrame.from_dict(films_info, orient ='index').transpose()
pd_films_info
```

## IMDb Scraper

```python
imdb_url = []
df = pd.read_excel('../data/pd_films_info.xlsx')

for u in df['IMDB_id']:
    if u == 'no_imdbid_found':
        imdb_url.append('')
    else:
        imdb_url.append('https://www.imdb.com/title/'+str(u)+'/') # combine the url link with each movie's id on imdb
print(len(imdb_url))
```

```python
headers = {
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36', # user-agent  # a relative quality factor. It specifies what language the user would prefer, on a scale of 0 to 1; Refs: https://www.w3.org/Protocols/rfc2616/rfc2616-sec14.html#sec14.4
    }
url = 'https://www.imdb.com/title/tt0399650/?ref_=fn_al_tt_1'
page = requests.get(url, headers = headers)
print(page)
```

```python
title_en = []
rate = []
reviews = []
year = []
boxOffice = []
budget = []

n = 1
for url in imdb_url:
    if url == '':
        title_en.append('')
        rate.append('')
        reviews.append('')
        year.append('')
        boxOffice.append('')
        budget.append('')
    else:
        # Pause the loop
        sleep(randint(3,6))
        headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
        page = requests.get(url, headers = headers)
        html_soup = BeautifulSoup(page.content, 'html.parser')

        try:
            title_1 = html_soup.find('div', class_='TitleBlock__TitleContainer-sc-1nlhx7j-1 jxsVNt')
            title_2 = title_1.find('h1').get_text()
            title_en.append(title_2)
        except:
            title_1 = 'no_title_found'
            title_en.append(title_1)

        ## rate
        try:
            rate_1 = html_soup.find('div', class_='AggregateRatingButton__ContentWrap-sc-1ll29m0-0 hmJkIS')
            rate_2 = rate_1.find('span', class_='AggregateRatingButton__RatingScore-sc-1ll29m0-1 iTLWoV').get_text()
            rate.append(rate_2)
        except:
            rate_1 = 'no_rate_found'
            rate.append(rate_1)

        ## reviews
        try:
            reviews_1 = html_soup.find('div', class_='AggregateRatingButton__TotalRatingAmount-sc-1ll29m0-3 jkCVKJ').get_text()
            reviews.append(reviews_1)
        except:
            reviews_1 = 'no_reviews_found'
            reviews.append(reviews_1)

        ## year
        try:
            year_1 = html_soup.find('div', class_='TitleBlock__TitleMetaDataContainer-sc-1nlhx7j-2 hWHMKr')
            year_2 = year_1.find('span', class_='TitleBlockMetaData__ListItemText-sc-12ein40-2 jedhex').get_text()
            year.append(year_2)
        except:
            year_1 = 'no_year_found'
            year.append(year_1)

        ## box office
        try:
            box_1 = html_soup.find('div', class_='TitleMainBelowTheFoldGroup__TitleMainPrimaryGroup-sc-1vpywau-1 btXiqv ipc-page-grid__item ipc-page-grid__item--span-2')
            box_2 = box_1.find('section', cel_widget_id="StaticFeature_BoxOffice")
            box_3 = box_2.find('ul', class_="ipc-metadata-list ipc-metadata-list--dividers-none ipc-metadata-list--compact BoxOffice__MetaDataListBoxOffice-sc-40s2pl-0 fxZYqi ipc-metadata-list--base")
            box_4 = box_3.find('span', class_='ipc-metadata-list-item__list-content-item').get_text()
            boxOffice.append(box_4)
        except:
            box_1 = 'no_box_found'
            boxOffice.append(box_1)

        ## budget
        try:
            bud_1 = html_soup.find('li', class_='ipc-metadata-list__item BoxOffice__MetaDataListItemBoxOffice-sc-40s2pl-2 gwNUHl')
            bud_2 = bud_1.find('span', class_='ipc-metadata-list-item__list-content-item').get_text()
            budget.append(bud_2)
        except:
            bud_1 = 'no_budget_found'
            budget.append(bud_1)
    
    print(n)
    n = n+1
```

```python
print(len(imdb_url))
print(len(title_en))
print(len(rate))
print(len(reviews))
print(len(year))
print(len(boxOffice))
print(len(budget))
```

## Reviews Scraper of IMDb

Scrape the reviews of the short-listed movies

```python
imdb_re_url = ['tt0299977', 'tt3781476', 'tt10627720', 'tt1362045', 'tt3336934', 'tt4019426', 'tt1717715', 'tt4862468', 'tt3594826', 'tt11952874', 'tt7587984', 'tt1825731']
imdb_re_title = ['英雄', '捉妖记', '哪吒之魔童降世', '白银帝国', '大明劫', '绣春刀', '西游记之大闹天宫', '大话西游3', '道士下山', '木兰：横空出世', '神探蒲松龄', '大笑江湖']
print(len(imdb_re_url))
print(len(imdb_re_title))
```

```python
headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
page = requests.get('https://www.imdb.com/title/tt10627720/reviews?sort=helpfulnessScore&dir=desc&ratingFilter=0', headers = headers)
html_soup = BeautifulSoup(page.content, 'html.parser')
```

```python
title = []
reviews_num = []
reviews_id = []
reviews_con = []
  
r = 0
for u in imdb_re_url:
    sleep(randint(3,6))
    headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
    url = 'https://www.imdb.com/title/'+str(u)+'/reviews?sort=helpfulnessScore&dir=desc&ratingFilter=0'
    page = requests.get(url, headers = headers)
    html_soup = BeautifulSoup(page.content, 'html.parser')
  
    reviews_1 = html_soup.find_all('div', class_='display-name-date')
    n = 1
    for re in reviews_1:
        title.append(imdb_re_title[r])
        reviews_num.append(n)
        id_ = re.find('a').get_text()
        reviews_id.append(id_)
        n = n+1

    reviews_2 = html_soup.find_all('div', class_='content')
    for re in reviews_2:
        con = re.find('div', class_='text show-more__control').get_text()
        reviews_con.append(con)
    print(r)
    r = r+1
    
print(len(title))
print(len(reviews_num))
print(len(reviews_id))
print(len(reviews_con))
print(title)
print(reviews_num)
print(reviews_id)
print(reviews_con)      
```

```python
films_review_imdb = {
    'title':title,
    'reviews_number':reviews_num,
    'reviews_id':reviews_id, 
    'reviews_content':reviews_con,
}
```

```python
films_review_imdb = pd.DataFrame.from_dict(films_review_imdb, orient ='index').transpose()
films_review_imdb
```
