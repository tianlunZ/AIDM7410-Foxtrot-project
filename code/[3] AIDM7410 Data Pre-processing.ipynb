```python
import requests 
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../data/pd_films_info.xlsx')
```

```python
df['Douban_type']
```

```python
gz = []
for t in df['Douban_type']:   ## check whether there is a keyword of costume in Douban_type
    types = t.split()
    #print(types)
    if len(types) == 1:
        if types[0] == '古装':
            gz.append('Yes')
        else:
            gz.append(' ')
    elif len(types) == 2:
        if types[0] == '古装':
            gz.append('Yes')
        elif types[1] == '古装':
            gz.append('Yes')
        else:
            gz.append(' ')
    elif len(types) == 3:
        if types[0] == '古装':
            gz.append('Yes')
        elif types[1] == '古装':
            gz.append('Yes')
        elif types[2] == '古装':
            gz.append('Yes')
        else:
            gz.append(' ')
    elif len(types) == 4:
        if types[0] == '古装':
            gz.append('Yes')
        elif types[1] == '古装':
            gz.append('Yes')
        elif types[2] == '古装':
            gz.append('Yes')
        elif types[3] == '古装':
            gz.append('Yes')
        else:
            gz.append(' ')
    else:
        if types[0] == '古装':
            gz.append('Yes')
        elif types[1] == '古装':
            gz.append('Yes')
        elif types[2] == '古装':
            gz.append('Yes')
        elif types[3] == '古装':
            gz.append('Yes')
        elif types[4] == '古装':
            gz.append('Yes')
        else:
            gz.append(' ')
print(gz)
```

```python
title = []
db_rate = []
imdb_rate = []
boxoffice = []
budget = []
year = []
type_ = []
douban_higher = []
imdb_re = []
imdb_url = []
douban_url = []
imdb_id = []

for t in df['title']:
    title.append(t)
for dbrate in df['Douban_rate']:
    db_rate.append(dbrate)
for imdbrate in df['IMDB_rate']:
    imdb_rate.append(imdbrate)
for bo in df['IMDB_boxoffice']:
    boxoffice.append(bo)
for bu in df['IMDB_budget']:
    budget.append(bu)
for y in df['year']:
    year.append(y)
for ty in df['Douban_type']:
    type_.append(ty)
for im in df['IMDB_reviews']:
    imdb_re.append(im)
for id_ in df['IMDB_id']:
    imdb_id.append(id_)
for iurl in df['IMDB_url']:
    imdb_url.append(iurl)
for durl in df['Douban_url']:
    douban_url.append(durl)
#rate = df[['Douban_rate', 'IMDB_rate']]
#rate
#for 
n = range(0, 715)
for i in n:
    try:
        if float(df['Douban_rate'][i]) > float(df['IMDB_rate'][i]):
            douban_higher.append('Yes')
        else:
            douban_higher.append('No')
    except:
        douban_higher.append('No')
print(len(douban_higher))
```

```python
films_gz = {
    'title': title,
    'year': year, 
    'guzhuang':gz,
    'type':type_,
    'Douban_higher':douban_higher,
    'Douban_rate': db_rate,
    'IMDB_rate':imdb_rate,
    'IMDB_reviews':imdb_re,
    'boxoffice':boxoffice,
    'budget':budget,
    'IMDB_url':imdb_url,
    'IMDB_id':imdb_id,
    'Douban_url':douban_url
}
```

```python
pd_films_gz = pd.DataFrame.from_dict(films_gz, orient ='index').transpose()
pd_films_gz
```

```python
pd_films_gzdy = pd_films_gz[pd_films_gz['guzhuang']=='Yes'] # select those movies with a keyword of costume
pd_films_gzdy   
```

```python
len(pd_films_gzdy)
```

```python
pd_films_fgzdy = pd_films_gz[pd_films_gz['guzhuang']==' ']
pd_films_fgzdy
```

```python
pd_films_gzdy.to_excel('pd_films_gzdy_2.xlsx')
```

```python
doubanhigh = pd_films_gzdy[pd_films_gzdy['Douban_rate']>pd_films_gzdy['IMDB_rate']]
print('古装电影中，豆瓣评分高于IMDB的数量：', len(doubanhigh))
print('古装电影总体数量：', len(pd_films_gzdy))
print('古装电影中，豆瓣评分高于IMDB的比例：', len(doubanhigh)/len(pd_films_gzdy))
```
